## <span style="font-family:Times New Roman"> Pre-processing the Data</span>
---

<span style="font-family:Times New Roman; font-size:1.25em">Before we analyse the data, we must clean the dataset to remove null values and duplicates. </span>

<span style="font-family:Times New Roman; font-size:1.25em">In order to remove the null values, we must first identify the columns that contain null values using the 'info()' function: </span>

```python
d1.info()
```

<span style="font-family:Times New Roman; font-size:1.25em">From the output, we can see that the columns that contain null values include </span>
    
<span style="font-family:Times New Roman; font-size:1.25em">
<li>fixed acidity
<li>citric acid
<li>chlorides
<li>free sulfur dioxide
<li>total sulfur dioxide
<li>density
<li>pH
<li>quality 
</span>

    
<span style="font-family:Times New Roman; font-size:1.25em">In order to remove the null values, I have decided to fill in the null values with the median values of the respective columns. </span>

<span style="font-family:Times New Roman; font-size:1.25em">I chose this method due to the fact that it would not remove any information connected to the rows that contain null values. </span>

<span style="font-family:Times New Roman; font-size:1.25em">I decided to fill the null cells with the median value as it is the most reliable measure of central tendency.</span>

```python
med_FA = d1['fixed acidity'].median()
med_CA = d1['citric acid'].median()
med_CL = d1['chlorides'].median()
med_FSD = d1['free sulfur dioxide'].median()
med_TSD = d1['total sulfur dioxide'].median()
med_DE = d1['density'].median()
med_PH = d1['pH'].median()
med_QU = d1['quality'].median()
```

```python
d1['fixed acidity'] = d1['fixed acidity'].fillna(med_FA)
d1['citric acid'] = d1['citric acid'].fillna(med_CA)
d1['chlorides'] = d1['chlorides'].fillna(med_CL)
d1['free sulfur dioxide'] = d1['free sulfur dioxide'].fillna(med_FSD)
d1['total sulfur dioxide'] = d1['total sulfur dioxide'].fillna(med_TSD)
d1['density'] = d1['density'].fillna(med_DE)
d1['pH'] = d1['pH'].fillna(med_PH)
d1['quality'] = d1['quality'].fillna(med_QU)
```

```python
d1
```

<span style="font-family:Times New Roman; font-size:1.25em">
The next step in cleaning the data is to remove any duplicates present.
</span>

```python
d1 = d1.drop_duplicates()
d1 
```

<span style="font-family:Times New Roman; font-size:1.25em">
From this step, it can be observed that there were no duplicates as the number of rows did not diminish.
